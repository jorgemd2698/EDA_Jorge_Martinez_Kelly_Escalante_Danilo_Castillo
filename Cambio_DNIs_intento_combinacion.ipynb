```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sqlite3
import random
import string
```

```python
# Definimos una funcion para reemplazar la informacion sensible de los datos, en este caso, los DNIs
def generar_dni():
    letras = "TRWAGMYFPDXBNJZSQVHLCKE"
    numero = random.randint(0, 99999999)
    letra = letras[numero % 23]
    return f"{numero:08d}{letra}"
# Evitamos que los DNIs coincidan
def generar_dnis_unicos(n):
    dnis = set()
    while len(dnis) < n:
        dnis.add(generar_dni())
    return list(dnis)
```

```python
df_accidentes = pd.read_csv("./accidentes.csv")
df_a = df_accidentes.copy()
df_a["NIF"] = generar_dnis_unicos(len(df_a))
df_a.info()
```

```python
# unimos ambos dataframes
df_licencias = pd.read_csv("./licencias_2024_limpio_P.csv")
df_l = df_licencias.copy()
# Aqui no tengo una ifnromacion en comun para poder combinarlos y desechar las columnas que nos sobran
df = pd.merge(df_l, df_a, on=["Nombre", "Apellido1"], how='inner')
df.info()
```
