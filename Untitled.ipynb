```python
import numpy as np
import pandas as pd
import re
import unicodedata

pd.options.mode.copy_on_write = True # CoW por defecto a partir de Pandas 3.0.0
```

```python
# carga de data set
df_raw = pd.read_csv(
    "./datasets/Licencias_2024.csv",
    low_memory=False
)

#Durante la carga de datos se detectaron columnas con tipos mixtos, particularmente en variables como el código postal. 
#Dado que estas variables no son numéricas analíticas, se trataron como categóricas y no afectaron el análisis posterior.
```

```python
# Normalización de nombres de columnas
def normalizar_texto(texto):
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFKD', texto)
    texto = texto.encode('ascii', 'ignore').decode('utf-8')
    texto = texto.replace(' ', '_')
    return texto

df.columns = [normalizar_texto(col) for col in df.columns]
```

```python
# Eliminar columnas duplicadas
cols_mayusculas = [
    'alta', 'apellido1', 'apellido2', 'bloqueado_envios',
    'categoria', 'codigopostaldomicilio', 'complementos',
    'cuaderno19', 'cuota_socio', 'descripcionentidad',
    'descripcionprovinciadomicilio', 'direcciondomicilio',
    'discapacidad', 'envio_tarjeta', 'estado',
    'fechaaltasolicitud', 'fechanacimiento', 'importeampliacion',
    'importecomplemento', 'importedescuento', 'importefmm',
    'importefederacionesp', 'importeseguro', 'modalidad',
    'niffamiliar', 'nsocio', 'nombre', 'numeroclub',
    'poblaciondomicilio', 'relacion', 'sexo', 'temporada',
    'tipotarjeta'
]

df = df.loc[:, ~df.columns.duplicated()]
```

```python
#Eliminación de columnas 100% nulas
nulos = df.isna().mean()
cols_100_nulas = nulos[nulos == 1].index

df.drop(columns=cols_100_nulas, inplace=True)
```

```python
# eliminar columnas no analiticas #Datos personales / sin valor estadístico

cols_eliminar = [
    'nombre',
    'apellido1',
    'apellido2',
    'direcciondomicilio',
    'niffamiliar',
    'tipotarjeta'
]

df.drop(columns=cols_eliminar, inplace=True, errors='ignore')
```

```python
# Eliminación de columnas sin valor analítico para el EDA

cols_eliminar_extra = [
    # Administrativas / logísticas
    'numeroclub',
    'descripcionentidad',
    'cuaderno19',
    'alta',
    'envio_tarjeta',
    'bloqueado_envios',
    'relacion',

    # Geográficas de alta granularidad
    'codigopostaldomicilio',
    'poblaciondomicilio',

    # Económicas no utilizadas directamente
    'importefmm',
    'importefederacionesp',
    'importecomplemento',
    'importedescuento',
    'importeampliacion'
]

df.drop(columns=cols_eliminar_extra, inplace=True, errors='ignore')

# Verificación
df.columns
#Se eliminaron variables administrativas, logísticas, geográficas de alta cardinalidad y económicas no directamente relacionadas con el análisis de
#siniestralidad, con el objetivo de reducir ruido y centrar el análisis en variables explicativas.

```

```python
# conversion de fechas 
df['fechanacimiento'] = pd.to_datetime(df['fechanacimiento'], errors='coerce')
df['fechaaltasolicitud'] = pd.to_datetime(df['fechaaltasolicitud'], errors='coerce')
```

```python
#crear variable edad 
df['edad'] = (
    pd.Timestamp('2024-12-31') - df['fechanacimiento']
).dt.days // 365
```

```python
# Limpieza y normalizacion de sexo # se detectó que la variable sexo estaba codificada en origen mediante las letras ‘H’ y ‘M’.
#Durante la exploración se corrigió una reasignación previa y se realizó una normalización semántica partiendo nuevamente de los datos originales
#para preservar la correcta distribución de la variable.

df['sexo'] = df_raw['Sexo'].copy()

# Normalización básica
df['sexo'] = (
    df['sexo']
    .astype(str)
    .str.strip()
    .str.upper()
)

# Mapeo SEMÁNTICO CORRECTO
df['sexo'] = df['sexo'].replace({
    'H': 'MASCULINO',
    'M': 'FEMENINO'
})
```

```python
#Conversión de variables categóricas   
cols_cat = [
    'sexo',
    'modalidad',
    'categoria',
    'discapacidad',
    'descripcionprovinciadomicilio',
    'poblaciondomicilio'
]

for col in cols_cat:
    if col in df.columns:
        df[col] = df[col].astype('category')

# se convierten a category para reflejar adecuadamente su naturaleza estadística, mejorar la eficiencia del procesamiento y facilitar el análisis exploratorio.
```

```python
# Análisis de cardinalidad

def resumen_cardinalidad(df):
    return (
        pd.DataFrame({
            'tipo_dato': df.dtypes,
            'valores_unicos': df.nunique(),
            'cardinalidad_%': round(df.nunique() / len(df) * 100, 2)
        })
        .sort_values('cardinalidad_%', ascending=False)
    )

cardinalidad = resumen_cardinalidad(df)
cardinalidad
```
